```python
import pandas as pd
```

```python
 # 1. The Raw Data (List of Dictionaries)
data = [
    {"name": "Harry", "score": 37.21, "grade": "B"},
    {"name": "Berry", "score": 37.21, "grade": "B"},
    {"name": "Tina", "score": 37.2, "grade": "C"},
    {"name": "Akriti", "score": 41.0, "grade": "A"},
    {"name": "Harsh", "score": 39.0, "grade": "B"}
]


# 2. The Magic Command: Convert to DataFrame
df = pd.DataFrame(data)

# 3. View the data
print(df)
```

```python
# Peeking Data
print(df.head())    # Prints first 5 rows
print(df.head(2))   # Prints first 2 rows
```

```python
# Selecting a single column
print(df["name"])
print(df["score"])
print(df["grade"])
```

```python
# Filtering data

# print(df["score"] > 37)

high_scores = df[["score"]] > 37
high_score = df[df["score"] > 37]

# print(high_scores)
print(high_score)

b_grade = df[df["grade"] == "B"]
print(b_grade)
```

```python
# Sorting
# df.sort_values(by = "score" , ascending=False)
df.sort_values(by = "score")
print(df.sort_values(by = "score", ascending=False))

print(df.sort_values(by = "name"))


# Pass LISTS to both arguments
df.sort_values(by=["score", "name"], ascending=[False, True])
```

```python
# Quick Stats (Aggregation)

print(df["score"].mean())  # Average score
print(df["score"].max())   # Highest score
print(df["score"].min())   # Lowest score
print(df["score"].sum())   # Total of all scores
```

```python
# Groupby

df.groupby("grade").mean(numeric_only=True)
```

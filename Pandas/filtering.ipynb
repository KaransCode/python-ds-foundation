Filtering in Pandas

```python
import pandas as pd
```

```python
pokemon_data = pd.read_csv("pokemon.csv")

# Filtering Tall Pokemon (based on Height)
tall_pokemon = pokemon_data[pokemon_data["Height"] >= 3]
print(tall_pokemon)

# Filtering Heavy Pokemon (based on Weight)
heavy_pokemon = pokemon_data[pokemon_data["Weight"] > 120]
print(heavy_pokemon)
num = heavy_pokemon["Name"].count()
print(num)
```

```python
# Finding Legendary Pokemons from the data sets
legendary_pokemon = pokemon_data[pokemon_data["Legendary"] == 1]
# print(legendary_pokemon) 

# Finding Water Pokemons from the data sets
water_pokemon = pokemon_data[(pokemon_data["Type1"] == "Water")]

# Or condition
# water_pokemon = pokemon_data[(pokemon_data["Type2"] == "Water") | (pokemon_data["Weight"] >= 100)]
# print(water_pokemon)

# And Condition
water_pokemon = pokemon_data[(pokemon_data["Type1"] == "Fire") & (pokemon_data["Height"] >= 1)]
print(water_pokemon)
```

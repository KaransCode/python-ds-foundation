```python
import pandas as pd

# Creating a Series
marks = pd.Series([78, 85, 90, 66, 88])
print("Series:\n", marks)

# Creating a DataFrame
data = {"Name": ["Piya", "Nitin", "Ravi", "Sumit", "Vikas"],
        "Marks": [78, 85, 90, 66, 88]}

df = pd.DataFrame(data)
print("\nDataFrame:\n", df)
```

```python
# Reading CSV file
df = pd.read_csv("students.csv")

# Basic inspection
print(df.head(1))
print()
print(df.tail(2))
print("\nShape:", df.shape)
print("\nColumns:", df.columns)
print("\nInfo:")
df.info()
```

```python
# Column selection
print(df["Name"])

# Conditional filtering
high_scorers = df[df["Score"] > 80]
print("\nStudents scoring above 80:\n", high_scorers)
```

```python
data = {"Subject": ["Math", "Math", "Python", "Python"],
         "Marks": [78, 85, 90, 88] }

df = pd.DataFrame(data)

# Aggregations
print("Average Marks:", df["Marks"].mean())
print("Maximum Marks:", df["Marks"].max())
print("Minimum Marks:", df["Marks"].min())

# GroupBy
grouped = df.groupby("Subject")["Marks"].mean()
print("\nAverage Marks by Subject:\n", grouped)
```

```python
import numpy as np
data = {"Name": ["Piya", "Nitin", "Karan", "Ravi", None],
        "Marks": [78, np.nan, 90, 70, 66]}

df = pd.DataFrame(data)
print("Original Data:\n", df)

# Handling missing values
print()
print(df.isna().sum())  # isna() 
df["Marks"].fillna(df["Marks"].mean(), inplace=True)
df.dropna(inplace=True)

print("\nCleaned Data:\n", df)
```

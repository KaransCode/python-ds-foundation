```python
import pandas as pd
```

Data cleaning = 
the process of fixing/ removing: 
incomplete, incorrect, or irrelevant data.
~ 75% of work done with Pandas is data cleaning

```python
df = pd.read_csv("pokemon.csv")
# 1. Drop irrelevant columns
drop_col = df.drop(columns=["Legendary", "No"])
# print(df)
# print(drop_col)

# 2. Handle missing data
# Drop NA Values or Fill NA Values
drop_row = df.dropna(subset=["Type2"])
# print(drop_row.count())
# print(drop_row.to_string())

fill_df = df.fillna({"Type2" : "None"})
# print(fill_df)
# print(fill_df.where("Type2" == None))

# 3. Fix Inconsistent Values

df["Type1"] = df["Type1"].replace({"Grass" : "GRASS", "Fire" : "FIRE"})
# print(df)

# 4. Standardized Text

# df["Name"] = df["Name"].str.lower()
df["Name"] = df["Name"].str.upper()
# print(df)


# 5. Fix Data Types

df["Legendary"] = df["Legendary"].astype(bool)
print(df)
```

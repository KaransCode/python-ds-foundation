Series in Pandas

```python
data = [101,102,201,21]
print(data)
print(type(data))

ob_data = ["A","B","C","D"]
o_series = pd.Series(ob_data)
# print(o_series)
# print(type(o_series))

series_data = pd.Series(data, index=["a","b","c","d"])
print(series_data)
print(type(series_data))
```

```python
# loc and iloc
# loc = location based on name of index - a,b,c or room1 , room2 etc.
print(series_data.loc["d"])
series_data.loc["b"] = 111
# print(series_data)

# iloc = integer-location 0,1,2 index based
print(series_data.iloc[0])

print("iloc : ")
for i in range(len(series_data)):
    print(series_data.iloc[i])
```

```python
print(series_data[series_data > 101])
```

```python
# Creating series from python dictionary

calories = { "Day1" : 1200 , "Day2" : 1500, "Day3" : 1700,
            "Day4" : 2100}
s_calories  = pd.Series(calories)
print(s_calories)
s_calories.loc["Day3"] += 300
print((s_calories.loc["Day3"]))
```

```python
print(s_calories)
print(s_calories[s_calories > 1500])
```

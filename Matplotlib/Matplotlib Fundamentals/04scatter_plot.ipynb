```python
# Importing essential libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# scatter graph Shows the relationship between two variables Helps to identify
# a correlation (+,-,None) Example: Study hours vs. Test scores

hr_study = np.random.randint(0,10,8)
marks = np.random.randint(33,60,8)

hr_study1 = np.random.randint(0,10,8)
marks1 = np.random.randint(33,60,8)

plt.scatter(hr_study, marks, color="skyblue", alpha=0.8, s=120 , linestyle="solid", label="MCD-1")
plt.scatter(hr_study1, marks1, color="purple", alpha=0.6, s=150, label="MCD-2")

plt.title("Hours Studies Vs. Marks scored")
plt.xlabel("Hours Study")
plt.ylabel("Marks Scored")

plt.legend()

plt.show()
```

```python
# Importing essential libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Creating random array from numpy
# x = np.random.randint(10,21,10)
# y = np.random.randint(30,100,10)

x = np.array([2020,2021,2022,2023,2024,2025])
y = np.array([10,15,21,31,59,99])

# Basic Plot 
plt.plot(x,y, marker="<")
plt.xlabel("Student's Age")
plt.ylabel("Student's Marks")
plt.show()
```

```python
# Labels 
# plt.title -  Title
# plt.xlabel - X - label 
# plt.ylabel - Y - label 

temperature = [20, 22, 24, 26, 28, 30, 32, 34, 36, 38]
sales = [150, 180, 210, 250, 310, 380, 460, 520, 590, 650]

# plt.plot(temperature, sales)
plt.plot(temperature, sales, marker='o', linestyle="dashdot", color='r', label='Ice Cream Sales')

plt.title('Ice Cream Sales vs Temperature', color="Black", fontsize=18, fontfamily="Arial", fontweight="bold")
plt.xlabel('Temperature (°C)', color="orange", fontsize=12, fontfamily="Arial")
plt.ylabel('Sales (Units)', color="Green", fontsize=12, fontfamily="Arial")

plt.tick_params(axis="both",color="magenta")
plt.xticks(temperature)

plt.grid(True)
plt.legend()

plt.show()
```

```python
# Customization in Plot
x = np.array([2020,2021,2022,2023,2024,2025])
y = np.array([10,15,21,31,59,99])
x1 = np.array([2020,2021,2022,2023,2024,2025])
y1 = np.array([1,11,28,39,65,79])

# customizing marker parameter 
# marker = ["o","<",">","^","."]
# markersize = 20 , 
# markerfacecolor (mfc) = [rgb color], 
# markeredgecolor (mec) = [rgb color]
plt.plot(x,y, marker="o" , markersize=12 , mfc="red" , mec="black")

# Linestyle parameter
# linestyle = ["dotted", "dashed", "solid", "dash-dot", "None" ]
# linewidth = 1,2,3,4,...
# color = [rgb color]
plt.plot(x1,y1, marker="<" , markersize=15 , mfc="green" , mec="blue" , linestyle= "dotted", linewidth=3, color="#71c2e5")
plt.xlabel("Student's Age")
plt.ylabel("Student's Marks")

plt.show()
```

```python
# We can use the same style again and again by storing it in a dict. and then calling it in plt.plot() function.
line_style = dict(marker="<" , markersize=15 , mfc="green" , mec="blue" , linestyle= "dotted", linewidth=3, color="#71c2e5")

plt.plot(y, **line_style)
# plt.grid(axis="both", color="grey", linewidth=0.5, linestyle="dashed")

plt.show()
```

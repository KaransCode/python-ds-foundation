```python
# Importing essential libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# bar chart = Plotting Categories and their values 

hobby = np.array(["Reading","Writing","Learning","Music","Traveling","Riding"])
level = np.array([7,4,6,9,8,6])

# Vertical bar chart
plt.bar(hobby, level, color=["lightblue","skyblue","blue"], linestyle="dashed")

# Horizontal bar chart
# plt.barh(hobby, level, color=["r","g","b","pink","cyan","magenta"], linestyle="dashed")

plt.title("Hobbies vs. My Interest", fontweight= "bold", fontsize=15)
plt.xlabel("Hobbies")
plt.ylabel("Interest Level")

plt.show()
```

```python
# Importing essential libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Histogram = A visual representation of the distribution of quantitative data.
#             They group values into bins (intervals)
#             and counts how many fall in leach range.

```python
scores = np.random.normal(loc=80,scale=10,size=100)
scores = np.clip(scores,0,100)

plt.hist(scores, bins=10, color="lightblue", ec="black")

plt.title("Student's Exams Scores")
plt.xlabel("Scores")
plt.ylabel("No. of Students")

plt.show()
```

```python
# Importing essential libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Pie chart = Circular Chart representing Categories and their proportion or percentage 

hobby = np.array(["Reading","Writing","Learning","Music","Traveling","Riding"])
level = np.array([7,4,6,9,8,6])

plt.pie( level, labels= hobby, autopct= "%1.2f%%", explode=[0,0,0,0.2,0,0], startangle=150, shadow=True)

plt.title("Hobbies vs. My Interest", fontweight= "bold", fontsize=15)

plt.show()
```

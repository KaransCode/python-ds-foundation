```python
import matplotlib.pyplot as plt
```

<h3> 1. Line Plot

```python

days = [1, 2, 3, 4, 5]
sales = [200, 250, 300, 280, 350]

plt.plot(days, sales)
plt.title("Sales Trend Over Days")
plt.xlabel("Days")
plt.ylabel("Sales")
plt.show()
```

<h3> Simple Line Chart Without any customization

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 30, 40, 50]

plt.figure(figsize=(6, 5))
plt.plot(x, y, marker='s', linewidth=2)
plt.title("Simple Line Plot with Grid", fontsize=14)
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python

days =  [1, 2, 3, 4, 5, 6, 7]
sales = [100, 220, 250, 280, 300, 310, 350]


plt.figure(figsize=(8, 5))

plt.plot(days, sales , marker="o", markersize=6 , markeredgecolor="seagreen", label="sales" , linewidth=2)
plt.title("Sales Trend Over Days" ,fontsize=12)
plt.xlabel("Days" , fontsize=12)
plt.ylabel("Sales", fontsize=12)

plt.xticks(color="red", fontsize=10)
plt.yticks(color="black", fontsize=10)

plt.grid(True, linestyle="--" , alpha=0.7)
plt.tight_layout()
plt.legend()

plt.show()
```

<h2>2. Bar Chart

```python
subjects = ["Math", "Physics", "Python", "Statistics", "DBMS" ]
marks =    [78, 85, 90, 60, 70]

plt.figure(figsize=(8, 5))

ax = plt.bar(subjects, marks, alpha=0.8)
plt.title("Marks by Subject", fontsize=14)

plt.bar_label(ax) # Marks Written on each subject bar

plt.xlabel("Subjects", color="seagreen", fontsize = 12)
plt.ylabel("Marks",color="steelblue", fontsize = 12)

plt.ylim(0, 100)
plt.tight_layout()
plt.show()
```

<h3>3. Scatter Plot

```python
hours_studied = [1, 2, 3, 4, 5 ,6]
scores = [40, 50, 60, 70, 85, 90]

plt.style.use('seaborn-v0_8-whitegrid')

plt.figure(figsize=(7,5))

plt.scatter(hours_studied, scores, s=80, alpha=0.8)
plt.title("Study Hours vs Score", fontsize=14)

plt.xlabel("Hours Studied")
plt.ylabel("Score")

plt.grid(True, linestyle=':')
plt.tight_layout()

plt.show()
```

<h3>4. Histogram

```python

marks = [45,46, 50, 51, 52 , 53 , 54, 55,55,56, 58, 60,61,62, 65, 70, 71,75, 79, 80,81, 85, 90]

plt.figure(figsize=(8, 6))

plt.style.context('ggplot')

plt.hist(marks, bins=6, edgecolor='black', alpha=0.6)
plt.title("Distribution of Marks", fontsize=14)

plt.xlabel("Marks")
plt.ylabel("Frequency")

plt.ylim(0,7)
plt.xlim(40,95)

plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

<h3>5. Pie Chart

```python
labels = ["Python", "Math", "Physics"]
time_spent = [40, 35, 25]

plt.figure(figsize=(4, 4))
plt.pie(time_spent, labels=labels, autopct='%1.1f%%', startangle=190, explode=[0.1,0,0])
plt.title("Time Spent on Subjects", fontsize=14)
plt.tight_layout()

plt.show()
```

<h3>6. Horizontal Bar Chart

```python

names = ["Piya", "Nitin", "Ravi", "Sumit", "Vikas"]
scores = [78, 85, 90, 66, 88]

plt.figure(figsize=(8, 5))

colors = ['steelblue', 'lightblue', 'seagreen', 'pink', 'orange', 'yellow']

plt.barh(names, scores, color = colors )
plt.title("Student Scores (Horizontal View)", fontsize=14)
plt.xlabel("Score")
plt.ylabel("Student")

plt.grid(axis='x', linestyle='--', alpha=0.6)

plt.tight_layout()

plt.show()
```

<h3>7. Box Plot

```python
marks = [45, 50, 55, 60, 65, 70, 75, 80, 90, 100]

plt.figure(figsize=(6, 5))
plt.boxplot(marks , patch_artist=True)
plt.title("Box Plot of Marks", fontsize=14)
plt.ylabel("Marks")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

7.1 Single Box Plot

```python

# Marks for different subjects
math_marks = [45, 50, 55, 60, 65, 70, 75, 80]
python_marks = [55, 60, 65, 70, 75, 85, 90, 95]
physics_marks = [40, 48, 52, 58, 63, 68, 72, 78]

# Combine data
data = [math_marks, python_marks, physics_marks]

plt.figure(figsize=(8, 5))

plt.boxplot(
    data,
    tick_labels=["Math", "Python", "Physics"],
    showmeans=True
)

plt.title("Box Plot of Marks by Subject", fontsize=14)
plt.ylabel("Marks")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

<h3>8. Multiple Lines on One Plot

```python
days = [1, 2, 3, 4, 5]
math = [70, 75, 80, 85, 90]
python = [65, 70, 78, 85, 95]

plt.figure(figsize=(8, 5))
plt.plot(days, math, marker='o', label="Math")
plt.plot(days, python, marker='s', label="Python")

plt.title("Performance Comparison Over Days", fontsize=14)
plt.xlabel("Days")
plt.ylabel("Marks")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

<h3>9. Subplots

```python
x = [1, 2, 3, 4, 5]
y1 = [10, 20, 30, 40, 50]
y2 = [5, 15, 25, 35, 45]

fig, axes = plt.subplots(1, 2, figsize=(10, 4))

axes[0].plot(x, y1, marker='o')
axes[0].set_title("Line Plot")

axes[1].bar(x, y2)
axes[1].set_title("Bar Plot")

plt.tight_layout()
plt.show()
```

<h3>10. Area Plot

```python

days = [1, 2, 3, 4, 5]
sales = [100, 150, 200, 250, 300]

plt.figure(figsize=(8, 5))
plt.plot(days,sales, color = "steelblue", marker="o")
plt.fill_between(days, sales, alpha=0.6)
plt.title("Cumulative Sales Trend", fontsize=14)
plt.xlabel("Days")
plt.ylabel("Sales")
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

<h3>11. Stack Plot

```python
days = [1, 2, 3, 4, 5]
python = [20, 25, 30, 35, 40]
math = [15, 20, 25, 30, 35]

plt.figure(figsize=(8, 5))
plt.stackplot(days, python, math, labels=["Python", "Math"], alpha=0.8)

plt.title("Subject-wise Study Time", fontsize=14)
plt.xlabel("Days")
plt.ylabel("Hours")

plt.legend(loc="upper left")
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```

<h3>12. Annotated Plot

```python

x = [1, 2, 3, 4, 5]
y = [50, 60, 70, 85, 90]

plt.figure(figsize=(7, 5))
plt.plot(x, y, marker='o')
plt.annotate("Highest Score", xy=(5, 90), xytext=(4, 80),
             arrowprops=dict(arrowstyle="->"))

plt.title("Performance Trend", fontsize=14)
plt.xlabel("Attempts")
plt.ylabel("Score")
plt.grid(True)
plt.tight_layout()
plt.show()
```

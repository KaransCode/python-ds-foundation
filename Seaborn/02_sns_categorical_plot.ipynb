```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
%reload_ext autoreload
%autoreload 2
```

# Categorical Plots #

# Bar Plots

```python
# Loading Datasets

# Car Crash Dataset
crash_df = sns.load_dataset('car_crashes')

crash_df.head()

# Tips Dataset
tips_df = sns.load_dataset('tips')
tips_df.head()
```

```python
sns.barplot(x='sex', y='total_bill', data=tips_df, estimator=np.median)
```

# Count Plot

```python
# sns.countplot( x='smoker', data=tips_df ,label="smoker")
# sns.countplot( x='sex', data=tips_df, label="sex")
# sns.countplot( x='time', data=tips_df, label="time")
sns.countplot( x='day', data=tips_df)
```

# Box Plot

```python
sns.boxplot(x='day', y='tip', data=tips_df, hue="sex")

plt.legend(loc=0)
```

# Violin Plot

Combination of Box plot and KDE Plot

```python
# sns.violinplot(data=tips_df , x='day', y='tip', hue='sex')

# sns.violinplot(data=tips_df , x='day', y='total_bill', hue='sex')

sns.violinplot(data=tips_df , x='day', y='total_bill', hue='sex', split=True)

plt.legend(loc=1)
```

# Strip Plot

```python
plt.figure(figsize=(10,5))
# sns.stripplot(data=tips_df , x='day', y='total_bill', jitter=True , hue="sex", palette='Blues')
sns.stripplot(data=tips_df , x='day', y='total_bill', jitter=True , hue="sex" , dodge=True)
```

# Swarm Plot

```python
sns.swarmplot(x='day', y='total_bill', data=tips_df, color="green")
```

# Palette

```python
plt.figure(figsize=(10,6))

sns.set_style('dark')
sns.set_context('talk')

sns.stripplot(x='day', y='total_bill', hue='sex' , data=tips_df, palette='afmhot')

plt.legend(loc=0)
```

```python
# Loading Datasets

# Car Crash Dataset
crash_df = sns.load_dataset('car_crashes')

# crash_df.head()
# crash_df.tail()

# Tips Dataset
tips_df = sns.load_dataset('tips')
tips_df.head()
```

# Distribution Plots

```python
sns.displot(crash_df["speeding"], kde=True, bins=9)

# KDE - Kernel Density Estimation (the curve showing in graph)
```

# Joint Plot

```python
# Regression
sns.jointplot( x='speeding', y='alcohol', data=crash_df, kind='reg') 

# Kde
sns.jointplot( x='speeding', y='alcohol', data=crash_df, kind='kde')

# Hexagon Distribution
sns.jointplot( x='speeding', y='alcohol', data=crash_df, kind='hex')
```

# KDE Plot

```python
sns.kdeplot(crash_df['speeding'])
```

# Pair Plot

```python
# sns.pairplot(crash_df) 

# sns.pairplot(tips_df, hue="sex")
sns.pairplot(tips_df, hue="sex", palette="Blues")
```

## Rug Plot

```python
# sns.rugplot(tips_df['total_bill'])
sns.rugplot(tips_df['tip'])
```

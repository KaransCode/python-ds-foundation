# PairGrid

```python
iris = sns.load_dataset('iris')
# species = iris.pop("species")
# iris
# iris_g = sns.PairGrid(iris, hue=species)
# iris_g.map_diag(plt.hist)
# iris_g.map_offdiag(plt.scatter)


tips_df_g = sns.PairGrid(tips_df, hue="sex")
tips_df_g.map_diag(plt.hist)
# tips_df_g.map_offdiag(plt.scatter)
tips_df_g.map_upper(plt.scatter)
tips_df_g.map_lower(sns.kdeplot)

tips_df_g = sns.PairGrid(tips_df , hue="sex" , x_vars=['smoker', 'day', 'time'] , y_vars=['total_bill', 'tip' , 'size'])
tips_df_g.map_diag(plt.hist)
tips_df_g.map_offdiag(sns.boxplot)
tips_df_g.add_legend()
plt.show()
```

# Facet Grid

```python
# facetg = sns.FacetGrid(tips_df , col="time", row="sex")
# facetg.map(plt.hist , 'total_bill', bins=5)

facetg = sns.FacetGrid(tips_df , col="time", row="smoker")
# facetg.map(plt.hist , 'total_bill', bins=5)

facetg.map(plt.scatter , 'total_bill', 'tip' , color = "orange")
plt.tight_layout()

facetg = sns.FacetGrid(tips_df , col="time", hue="smoker" , height=4 , aspect=1.2 , palette='Set1')
facetg.map(plt.scatter , 'total_bill' , 'tip' , ec='black')
facetg.add_legend()
```

```python
kws = dict(s = 50, linewidth=.5, edgecolor = 'w' )

tips_fg = sns.FacetGrid(tips_df, col= 'sex' , hue= 'smoker' , height=4, aspect=1.3, hue_order= ['Yes' , 'No'] , hue_kws= dict(marker=['^', 'v']))

tips_fg.map( plt.scatter, 'total_bill' , 'tip' , **kws)
```

```python
plt.figure(figsize=(8,8))
att_df = sns.load_dataset('attention')

att_df
att_fg = sns.FacetGrid(att_df , col="subject" , col_wrap=5 , height=1.5)
att_fg.map(plt.plot , 'solutions', 'score' , marker='o' , color='lightblue')
```
